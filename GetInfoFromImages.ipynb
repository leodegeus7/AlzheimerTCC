```python
import numpy as np
import shutil
import os
import fnmatch

import pandas as pd

names = ["" for x in range(0)]
fullnames = ["" for x in range(0)]
group = ["" for x in range(0)]
separation  = ["" for x in range(0)]
for path,dirs,files in os.walk('.'):
    for f in fnmatch.filter(files,'*.jpg'):
        fullname = os.path.abspath(os.path.join(path,f))
        last = fullname.split("/")[-1]
        if last[0] == "A":
            name = "_".join(last.split("_")[1:4])
        else:
            name = last.split(".")[0]
        fullnames.append(fullname)
        names.append(name)
        group.append(fullname.split("/")[-2])
        separation = fullname.split("/")[-3]
data = pd.DataFrame(data={"fullnames":fullnames,"names":names,"group":group,"separation":separation})
names = ["" for x in range(0)]
for i in data.names:
    name = i
    if len(i.split("_")) == 4:
        name = "_".join(i.split("_")[0:3])
    names.append(name)
    
data.names = names
data = data.drop_duplicates(subset=["names"])
```
